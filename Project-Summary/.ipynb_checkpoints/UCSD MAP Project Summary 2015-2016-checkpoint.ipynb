## Vocabulary to Understand
R-peak and RR Intervals: In a electrocardiography graph, we call the peaks R-peaks and the distance between consecutive R-peaks RR intervals. An RRITS is an RR interval time series, which is just a list of many RR intervals over a period of time.

ECG sensors: Sensors that can provide RR signal by measuring the electric signal generated from the expansion and contraction of heart chamber. They are the most accurate heart rate sensors.

PPG sensors: Light-based sensors that provide an RR signal by detecting the color change caused by blood flow in a body part. A variety of these devices exist, and even a smartphone camera can serve as a PPG sensor. For example, if a person places their finger on the flashlight of a smartphone, the tip of the finger would exhibit changes in redness as blood flow fluctuates. The camera would then record this change in a video.


## Abstract
The focus of our research is on the computation of RR interval time series. In the first part of our work, we extracted the RR intervals from the videos of a fingertip. We explored means of extracting RR intervals by converting each frame of the video into a representative number (average color intensity) that can be quantitatively analyzed. This is the principle at the basis of the PPG (Photoplethysmography) signal analysis.  In the second part of our work, we aligned the RR interval time series that we obtained previously with another longer RR interval time series (from the same subject and recorded at the same time) obtained from an ECG (Electrocardiography) sensor. We then aligned each PPG-derived RRITS with the ECG-derived RRITS by minimizing the absolute value of their distance.

```python
#returns whether or not the method was successful, and the average color value for RGB
#for an individual frame
def readFrame():
    hasFrame, frame = video.read();
    if not hasFrame:
        return False, 0, 0, 0
    totalRed = 0
    totalGreen = 0
    totalBlue = 0
    for col in range(0, width):
        # adds up all of the color values for each pixel in the frame
        # returns average value per pixel
        for row in range(0, height):
            totalRed += frame[row][col][2]
            totalGreen += frame[row][col][1]
            totalBlue += frame[row][col][0]
    return True, totalRed/1.0/width/height, totalGreen/1.0/width/height, totalBlue/1.0/width/height
    
```

#### Obtaining the Peaks in Color
Looking at the graphs above, each obvious peak in color is a heartbeat. Therefore, the difference in time between each peak is an RR interval. However, how do we determine which peaks are heartbeats? We can tell by looking at the graph which peaks are very likely heartbeats, but there are some mini-peaks (usually near the bottoms of the major peaks) that prevent us from just considering every local maximum a heartbeat peak. Therefore, we must test a particular threshold of frames before and after the considered peak to see if it is truly a heartbeat peak (which will be referred to as an R-peak). Below is my attempt to figure out the best frame threshold of a particular sample video in order to extract the correct peaks.

```python
#will return a list of the indexes of the peaks
def peakList(colorList, frameThreshold):
    peaks = []
    for i in range(1,len(colorList)-1):
        #check to see if greater than immediate surroundings
        if colorList[i-1] <  colorList[i]  and colorList[i+1] <  colorList[i]:
            #if so, check every other frame in the threshold
            for j in range(2, frameThreshold+1):
                #check everything in threshold before - will break if not a peak
                if i-j>0:
                    if colorList[i]<colorList[i-j]:
                        break
                #check everything in threshold after
                if i+j<len(colorList):
                    if colorList[i]<colorList[i+j]:
                        break
            else:
                peaks.append(i)
    return peaks
```

```python
r,g,b = readFromFile("data-analysis-lectures/color-values-PPG1.txt")
```

```python
#gets list of rr intervals when given a list of peak indexes, which is done in the peakList method
def rrList(peakIndexes):
    rrlist = [];
    for i in range(1,len(peakIndexes)):
        rr = (peakIndexes[i]-peakIndexes[i-1])*1/fps
        rrlist.append(rr)
    return rrlist
```

```python
#places each rr interval at its correct point in time (mainly for graphing purposes)
def timeList(rrList):
    times = [rrList[0]]
    for i in range(1,len(rrList)):
        times.append(times[i-1]+rrList[i])
    return times
```

```python
redRRList = rrList(peakList(r,9))
redRR = pd.Series(redRRList,index = timeList(redRRList))
%matplotlib inline
print "There are ", str(len(peakList(r,9))), "peaks."
redRR.plot()
```

#### Finding the best threshold
Changing the threshold of frames can change the accuracy of the peak identification, and therefore the accuracy of the RR intervals. Therefore, we processed multiple videos with the goal of finding the most accurate threshold for each color. 

```python
#the videos that will be tested
videoList = ["PPG-Analysis/VID_20150123_172926_F.mp4.txt",
    "PPG-Analysis/VID_20150123_173259_Q.mp4.txt",
    "PPG-Analysis/VID_20150123_173329_F.mp4.txt",
    "PPG-Analysis/VID_20150123_173358_Q.mp4.txt",
    "PPG-Analysis/VID_20150123_173427_F.mp4.txt",
    "PPG-Analysis/VID_20150123_173457_Q.mp4.txt",
    "PPG-Analysis/VID_20150123_173556_Q.mp4.txt",
    "PPG-Analysis/VID_20150123_173629_F.mp4.txt",
    "PPG-Analysis/VID_20150123_173700_Q.mp4.txt",
    "PPG-Analysis/VID_20150123_173757_Q.mp4.txt",
    "PPG-Analysis/VID_20150123_173028_F.mp4.txt",
    "PPG-Analysis/VID_20150123_172857_Q.mp4.txt"]
```

```python
thresholds = range(1,31)
```

```python
def getError(rrList):
    errors = 0
    for rr in rrList:
        #heart beat ranges from 50 to 110 pulses per minute
            if rr<.54 or rr>1.2:
                errors = errors+1
    return errors
```

```python
#goes through all vids, constructs matrix with vid,colors,errors
allErrors = []
for eachVideo in range(0,len(videoList)):
    allErrors.append([[],[],[]])
    
        
for nameIndex in range(0,len(videoList)):
    #for every vid, get colors
    r,g,b = readFromFile(videoList[nameIndex])
    colors = [r,g,b]
    for c in range(0,len(colors)):
        #list of errors for this paricular color and video
        errors = [0]*len(thresholds)
        for i in range(0,len(thresholds)):
        #for every vid, test each threshold for errors
        #add errors to approprate place on redErrors list
            rr = rrList(peakList(colors[c],i+1))
            if len(rr)>0:
                #relative error
                errors[i] = errors[i]+getError(rr)/(len(rr)*1.0)
            if len(rr)==0:
                #no rr intervals detected, which is obviously wrong, so 100% error
                errors[i] = errors[i]+1
        allErrors[nameIndex][c]=errors
#get average number of errors per video
reds= [0]*30
greens= [0]*30
blues = [0]*30
for a in range(0,len(allErrors)):
    for e in range(0,len(allErrors[a][0])):
            reds[e] = reds[e]+allErrors[a][0][e]
            greens[e] = greens[e]+allErrors[a][1][e]
            blues[e]= blues[e]+allErrors[a][2][e]

for i in range(0,len(reds)):
    reds[i] = reds[i]/(len(videoList)*1.0)
    greens[i] = greens[i]/(len(videoList)*1.0)
    blues[i] = blues[i]/(len(videoList)*1.0)
```

```python
#total threshold window
#for example, if it checks 7 frames foward and back, window is 14 frames
for i in range(0,len(thresholds)):
    thresholds[i]= thresholds[i]*2
```

In the following cells, I will categorize each video by quality. Quality is determined by the the average relative error when a video is analyzed with the a color's optimal frame threshold. For example, a video an average relative error of .2 (when a particular color/ opimal frame threshold is chosen) is of higher quality than a video of error .6. 

I will begin this process by categorizing the videos in a list where list[0]= videos with errors ranging from 0-.2, list[1] = errors from .2-.4, etc.

```python
def getQualityList(interval, videoList):
    #make sure 1/interval is a whole number
    if 1/interval!= int(1/interval):
        return []
    
    qList = []
    i=0
    index=0
    while(i<.9999):
        intervalList = []
        for video in range(0,len(videoList)):
            # the 6 means we're using the ideal threshold for red, which is 7 frames foward and back 
            #(this error is held at 6 since 0 holds the error for a threshold of 1, 1 holds a threshold of 2, etc.)
            if allErrors[video][0][6]>=i and allErrors[video][0][6]<i+interval:
                intervalList.append(video)
        qList.append(intervalList)
        i+=interval
    return qList
```

#### Matching video segments with full RR interval sets

Finally, now that we have these videos sorted by quality, we can try to match the ones that are of a high enough quality with their correct place in one of the two full RR interval sets from the ECG sensors (if by this point you forget what this is, refer back to "the Scenario" at the top of the page). Let's see how accurate the results were.

```python
#read in the signals
sig1 = []
sig2 = []
signal1 = open("PPG-Analysis/FullSignal1")
signal2 = open("PPG-Analysis/FullSignal2")
filecontents = signal1.readlines()
for line in filecontents:
    num = line.strip('\n')
    sig1.append((int)(num))
    
filecontents = signal2.readlines()
for line in filecontents:
    num = line.strip('\n')
    sig2.append((int)(num))
    
#convert from milliseconds to seconds
for i in range(0,len(sig1)):
    sig1[i]= sig1[i]/1000.0
for i in range(0,len(sig2)):
    sig2[i]= sig2[i]/1000.0
```

```python
#returns distance between video segment and a particular portion in the full signal
def getDifference(segment,fullSignal, beginning):
    relDist = 0
    divisor=0
    for i in range(0,len(segment)):
        relDist += abs(segment[i]-fullSignal[i+beginning])
        divisor+=(fullSignal[i+beginning])
    relDist = relDist/divisor
    return relDist
    
```

```python
#we've decided to obtain rr intervals at a total frame threshold of 14 (7 foward and back) and using the color red
reds = []
for i in range(0,len(videoList)):
    r,g,b=readFromFile(videoList[i])
    reds.append(r)
frameThresh = 7

rrLists = []
for i in range(0,len(reds)):
    rrLists.append(rrList(peakList(reds[i],7)))
```

```python
def findBestMatch(segment,fullSignal):
    diffs = []
    for e in range(0,len(fullSignal)-len(segment)+1):
        diffs.append(getDifference(segment,fullSignal,e))
    return diffs.index(min(diffs)),min(diffs)
```

```python
#Show the similarity of the best match
print findBestMatch(rrLists[1],sig1)
print findBestMatch(rrLists[1],sig2)
print rrLists[1]
print sig2[289:289+len(rrLists[1])]
```

Video 1 best matches starting at RR interval 289 in ECG signal (we'll later be converting this to seconds). The if you observe the 2 arrays above, you can see how similar their respective RR intervals are.

```python
#Videos shown to have the least error:
bestVideos = [1, 3, 5, 6, 8, 9, 10, 14, 17]

bestMatches = []
#contains lists that are defined by [segment,fullSignal,index,diffValue]
for video in bestVideos:
    match1,diffValue1 = findBestMatch(rrLists[video],sig1)
    match2,diffValue2 = findBestMatch(rrLists[video],sig2)
    
    if diffValue1<diffValue2:
        bestMatches.append([video, videoList[video],"Signal 1",match1,diffValue1])
    else:
        bestMatches.append([video, videoList[video],"Signal 2",match2,diffValue2])
for match in bestMatches:
    print match
```

```python
 #trying with videos of slightly lesser quality, but still decent
goodVideos =  [0, 11, 13, 15, 16,18]
goodMatches = []
#contains lists that are defined by [segment, segment name,fullSignal,index,diffValue]
for video in goodVideos:
    match1,diffValue1 = findBestMatch(rrLists[video],sig1)
    match2,diffValue2 = findBestMatch(rrLists[video],sig2)
    
    if diffValue1<diffValue2:
        goodMatches.append([video, videoList[video],"Signal 1",match1,diffValue1])
    else:
        goodMatches.append([video,videoList[video],"Signal 2",match2,diffValue2])
for match in goodMatches:
    print match
```

As you can see, any video ending in F matched to Signal 1 and any video ending in Q matched to Signal 2, so every video found its correct match!

```python
#getting the time for the placement of the video
def placementInSeconds(signal,index):
    seconds = 0
    for i in range(0,index):
        seconds += signal[i]
    return seconds
```

```python
#putting matches in seconds
allMatches = bestMatches+goodMatches
sig1Matches = []
sig2Matches = []
for i in range(0,len(allMatches)):
    if allMatches[i][2]=="Signal 1":
        sig1Matches.append(allMatches[i])
    else:
        sig2Matches.append(allMatches[i])
```
